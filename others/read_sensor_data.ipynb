```python
import matplotlib.pyplot as plt
import numpy as np

# file in this format ---->>>>     X Y
file = open("humidity.txt", "r") 
```

```python
X = []
Y = []

for line in file:
    a, b = line.split()
    Y.append(float(a))
    X.append(float(b))
    
print(len(Y), len(X))
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
print(Y)
```

```python
plt.plot(X,Y,  linewidth=2)
plt.title("Humidity Sensor")
plt.xlabel("Time in minutes(mins)")
plt.ylabel("Amplitude in percentage(%)")
plt.show()
```
