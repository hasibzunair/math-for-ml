```python
from keras.models import Sequential
from keras.layers import Dense
import numpy 
numpy.random.seed(7)
```

```python
dataset = numpy.loadtxt("diabetes.csv", delimiter=",", skiprows=1)
#input and output
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
print(X.shape)
print(Y.shape)
```

```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu')) #relu works best in most cases
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid')) #since output is 0 or 1
```

```python
'''Training - finding best set of weights to make predictions
    + Loss function to evaluate set of weights
    + Optimizer to search through different weights for network
'''
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
model.fit(X, Y, epochs=10, batch_size= 10, verbose=1)
```

```python
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
predictions = model.predict(X)
rounded = [round(x[0]) for x in predictions]
print(rounded)
```
